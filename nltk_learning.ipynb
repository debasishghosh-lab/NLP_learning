```python
import nltk
```

This cell imports the `nltk` library, making its functionalities available for use in the subsequent cells. It's a standard practice to import all required libraries at the beginning of the notebook.

```python
from nltk.stem import PorterStemmer
from nltk.corpus import stopwords
```

This cell imports `PorterStemmer` for stemming (reducing words to their root form) and `stopwords` from `nltk.corpus` for removing common words that do not add significant meaning to the text. These are crucial components for text cleaning.

```python
## Tokenization--> converts paragraph-sentences-words
nltk.download('punkt')
nltk.download('punkt_tab')
sentences=nltk.sent_tokenize(paragraph)

```

This cell performs **tokenization**, a process that splits the `paragraph` into individual sentences. It first downloads the necessary `punkt` and `punkt_tab` tokenizers (if not already present) and then uses `nltk.sent_tokenize()` to achieve this. The result is stored in the `sentences` list.

```python
sentences
```

This cell displays the `sentences` list, showing the individual sentences extracted from the original paragraph after tokenization. This helps to verify the output of the `sent_tokenize` function.

```python
import re
```

This cell imports the `re` module, which provides regular expression operations. Regular expressions are powerful for pattern matching and text manipulation, often used in text cleaning processes to remove unwanted characters.

```python
corpus=[]
for i in range(len(sentences)):
  review=re.sub('[^a-zA-Z]',' ',sentences[i])
  review=review.lower()
 # review=review.split()
  corpus.append(review)
corpus
```

This cell initiates the **text cleaning** process. It iterates through each sentence, removes non-alphabetic characters using `re.sub()`, converts the text to lowercase, and stores the cleaned sentences in a new list called `corpus`. This step prepares the text for further linguistic analysis.

```python
## Stemming
stemmer=PorterStemmer()
nltk.download('stopwords')
for i in corpus:
  words=nltk.word_tokenize(i)
  for word in words:
    if word not in set(stopwords.words('english')):
      print(stemmer.stem(word))
```

This cell demonstrates **stemming** using the `PorterStemmer`. It downloads the `stopwords` corpus (if not already downloaded) and then iterates through each sentence in the `corpus`. For each word, it checks if it's a stopword, and if not, it prints its stemmed form (e.g., 'averaging' becomes 'averag'). This reduces words to their root form.

This cell is currently empty. It can be used for additional imports or setup if needed later.

```python
## Lemmatization
from nltk.stem import WordNetLemmatizer
nltk.download('wordnet')

lemmatizer=WordNetLemmatizer()
nltk.download('stopwords')

for i in corpus:
  words=nltk.word_tokenize(i)
  for word in words:
    if word not in set(stopwords.words('english')):
      print(lemmatizer.lemmatize(word))
```

This cell demonstrates **lemmatization** using `WordNetLemmatizer`. It downloads the `wordnet` corpus and iterates through the cleaned sentences. Similar to stemming, it removes stopwords, but instead of just chopping off suffixes, lemmatization reduces words to their base or dictionary form (lemma), considering the word's meaning (e.g., 'averaging' becomes 'average').

```python
# Apply stopwords .lemmatize

corpus=[]
for i in range(len(sentences)):
  review=re.sub('[^a-zA-Z]',' ',sentences[i])
  review=review.lower()
  review=review.split()
  review=[lemmatizer.lemmatize(word) for word in review if not word in set(stopwords.words('english'))]
  review=' '.join(review)
  corpus.append(review)
corpus
```

This cell applies both **stopword removal and lemmatization** to the entire `sentences` list and updates the `corpus`. It first cleans each sentence, splits it into words, and then applies the `lemmatizer` only to non-stopwords. Finally, it joins the lemmatized words back into a string, creating a fully preprocessed `corpus` for further analysis.

```python
from sklearn.feature_extraction.text import CountVectorizer
cv=CountVectorizer()
X=cv.fit_transform(corpus)
```

This cell initializes a `CountVectorizer` from `sklearn.feature_extraction.text` and applies it to the preprocessed `corpus`. The `CountVectorizer` converts a collection of text documents to a matrix of token counts (term-document matrix), which is a common step for transforming text data into a numerical format suitable for machine learning algorithms.

```python
cv.vocabulary_
```
